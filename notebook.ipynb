```
import numpy as np
```

```
# Import Function 
def import_text(file_name, data_type='s'):
    """ For integers flag with i"""
    with open(file_name) as f:
        if data_type.lower() == 'i':
            return [int(line.rstrip()) for line in f]     
        elif data_type.lower() == 's':
             return [line.rstrip() for line in f]
    return None 
```

### December 1st

https://adventofcode.com/2020/day/1

```
# Part 1: Find the two entries that sum to 2020; what do you get if you multiply them together?
expenses = import_text('Dec1.txt', 'i')

# Assumes only one combination... 
np.prod(np.array([i for i in expenses if 2020 - i in expenses]))
```

```
# Part 2: What is the product of the three entries that sum to 2020?
results = []
for i in expenses:
    remainder = 2020 - i 
    for j in expenses:
        if j < remainder:
            if 2020 - (j + i) in expenses:
                results.append((i, j, 2020 - (j + i)))

np.prod(np.array(results[0]))
```

#### December 2nd

https://adventofcode.com/2020/day/2

```
# Part One: How many passwords are valid according to their policies?
passwords = import_text('Dec2.txt')

c = 0
for entry in passwords:
    policy, letter, password = entry.replace(':','').split()
    policy = policy.split('-')
    if password.count(letter) in range(int(policy[0]),int(policy[1]) + 1):
        c += 1
print(c)    
```

```
# Part Two: 
c = 0 
for entry in passwords:
    policy, letter, password = entry.replace(':','').split()
    policy = [int(i) - 1 for i in policy.split('-')]
    check = list(map(password.__getitem__, policy))
    if check.count(letter) == 1:
        c += 1
print(c)
```

#### Dec 3rd 
https://adventofcode.com/2020/day/3

```
trees = import_text('Dec3.txt')

# Use this function for both problems:

def slope_checker(right, down):
    position = 0
    row = 0
    no_trees = 0
    while row < len(trees):
        if trees[row][position % len(trees[row])] == "#":
            no_trees += 1
        position += right
        row += down
    return no_trees
```

```
# Part 1: Starting at the top-left corner of your map and following a slope of right 3 and down 1, how many trees would you encounter?
print(slope_checker(3,1))
# part 2 :
routes = [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]
print(np.product([slope_checker(*i) for i in routes]))
```
